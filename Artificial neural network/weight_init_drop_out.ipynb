# **Weight Initialization**

Weight initialization plays a crucial role in training deep neural networks. Proper initialization helps in stable gradient flow, faster convergence, and prevents vanishing or exploding gradients.


- ### **Criteria of Good Weight Initialization**

A good weight initialization technique should satisfy the following:

**Goals**
- Prevent vanishing gradients
- Prevent exploding gradients
- Maintain variance of activations across layers
- Maintain variance of gradients during backpropagation
- Break symmetry (weights should not be initialized to same values)

If weights are:
- Too small → gradients vanish
- Too large → gradients explode


---

- ### **Uniform Initialization**

In uniform initialization, weights are sampled from a uniform distribution within a specific range.

**General Form**

$$
w \sim U(-a, a)
$$

Where:
- $ a $ defines the range of distribution

Common simple choice:

$$
a = \frac{1}{\sqrt{n}}
$$

Where:
- $ n = number\ of\ input\ units\ (fan_{in}) $

**Advantages**
- Simple to implement
- Breaks symmetry

**Disadvantages**
- Not ideal for deep networks
- Can still lead to vanishing/exploding gradients


- ### **Xavier / Glorot Initialization**

Xavier initialization (also known as Glorot initialization) is designed to keep the variance of activations same across layers. It works well with Sigmoid and Tanh activation functions.

Let:

$$
fan_{in} = number\ of\ input\ units
$$

$$
fan_{out} = number\ of\ output\ units
$$


#### **Xavier Uniform**

Weights are sampled from:

$$
w \sim U \left(-\sqrt{\frac{6}{fan_{in} + fan_{out}}}, \sqrt{\frac{6}{fan_{in} + fan_{out}}} \right)
$$

#### **Xavier Normal**

Weights are sampled from:

$$
w \sim N \left(0, \frac{2}{fan_{in} + fan_{out}} \right)
$$

**Advantages**
- Maintains stable variance across layers
- Reduces vanishing/exploding gradients
- Suitable for Sigmoid/Tanh activations

**Disadvantages**
- Not ideal for ReLU-based networks


- ### **Kaiming He Initialization**

He initialization (also called Kaiming initialization) is specifically designed for ReLU and its variants. Since ReLU zeroes out half of the activations, He initialization compensates by adjusting variance.

Let:

$$
fan_{in} = number\ of\ input\ units
$$


#### **He Uniform**

Weights are sampled from:

$$
w \sim U \left(-\sqrt{\frac{6}{fan_{in}}}, \sqrt{\frac{6}{fan_{in}}} \right)
$$


#### **He Normal**

Weights are sampled from:

$$
w \sim N \left(0, \frac{2}{fan_{in}} \right)
$$

**Advantages**
- Designed for ReLU family activations
- Prevents vanishing gradients in deep networks
- Enables faster convergence

**Disadvantages**
- Not optimal for Sigmoid/Tanh activations

# **Dropout Layer**

Dropout is a regularization technique used in deep learning to reduce overfitting. During training, it randomly deactivates (drops) a fraction of neurons in a layer so that the network does not become overly dependent on specific neurons.

Instead of training one large neural network, dropout can be thought of as training many smaller subnetworks and averaging their predictions.


---

- ### **What is Dropout?**

Dropout randomly sets a fraction of neuron outputs to zero during training.

If:

$$
p = probability\ of\ dropping\ a\ neuron
$$

Then:

$$
1 - p = probability\ of\ keeping\ a\ neuron
$$

Each neuron is independently dropped with probability $$ p $$.


- ### **Why is Dropout Important?**

**Prevents Overfitting**
- Forces the network to learn more robust features.
- Prevents co-adaptation (neurons relying too much on specific other neurons).

**Acts as Regularization**
- Similar effect to training multiple models and averaging them.
- Improves generalization on unseen data.

**Reduces Model Complexity During Training**
- Makes the network less sensitive to noise in training data.


- ### **How Dropout Works**

The number of neurons in a layer is multiplied by it's $p$ value. 

For eg, 
if $p = 0.5$, and the number of neurons are 4, then 2 neurons will be deactivated in an epochs, the deactavation is completely random.

### **During Inference (Testing Phase)**

- No neurons are dropped.
- Full network is used.
- The wieghts are multiplied with the $p$ value.

**Summary**

- $ p $ = probability of dropping a neuron  
- Higher $$ p $$ → stronger regularization  
- Typical values: $$ p = 0.2 $$ to $$ 0.5 $$
