# **Activation Functions**
Activation functions are essential, non-linear, and differentiable mathematical equations applied to neural network outputs, allowing models to learn complex data patterns. By determining if a neuron should activate based on input, they prevent networks from behaving as simple linear regressions, with common types including ReLU, Sigmoid, Tanh, and Softmax.

## **Sigmoid**
The sigmoid function, specifically the logistic function, is a mathematical function that maps any real-valued number into a narrow range between 0 and 1. Its distinctive "S"-shaped curve makes it a cornerstone of machine learning, particularly for modeling probabilities.

$$ σ(z) = \frac{1}{1 + e^{-z}} $$
