# **Activation Functions**
Activation functions are essential, non-linear, and differentiable mathematical equations applied to neural network outputs, allowing models to learn complex data patterns. By determining if a neuron should activate based on input, they prevent networks from behaving as simple linear regressions, with common types including ReLU, Sigmoid, Tanh, and Softmax.

## **Sigmoid**
The sigmoid function, specifically the logistic function, is a mathematical function that maps any real-valued number into a narrow range between 0 and 1. Its distinctive "S"-shaped curve makes it a cornerstone of machine learning, particularly for modeling probabilities.

$$ σ(z) = \frac{1}{1 + e^{-z}} $$

![](https://media.geeksforgeeks.org/wp-content/uploads/20250131185746649092/Sigmoid-Activation-Function.png)

## **tanh**
The tanh (hyperbolic tangent) activation function is a non-linear, S-shaped function that maps input values to a range between -1 and 1. It is zero-centered, which often leads to faster convergence and more efficient training compared to the sigmoid function. It is widely used in hidden layers of neural networks, particularly RNNs, but can suffer from vanishing gradients.

$$ tanh(z) = \frac{e^{z} – e^{-z}}{e^{z} + e^{-z}} $$

![](https://media.geeksforgeeks.org/wp-content/uploads/20250214171817652462/tanh.png)

> The above two functions are prone to **Vanishing gradient problem**

## **ReLU**
The Rectified Linear Unit (ReLU) is the default activation function for deep neural networks. It outputs input directly if positive, otherwise zero, offering high computational efficiency and solving the vanishing gradient problem. It is widely used in CNNs and multi-layer perceptrons.
One problem with this activation function is that Neurons can become "stuck" at 0 if they only receive negative inputs, effectively stopping learning. Also, it isn't zero centric.

$$ ReLU(z) = max(0, z) $$

![](https://media.geeksforgeeks.org/wp-content/uploads/20250129162127770664/Relu-activation-function.png)

## **Leaky ReLU**
Leaky ReLU is an activation function for neural networks that addresses the "dying ReLU" problem by allowing a small, non-zero gradient for negative inputs. It prevents neurons from becoming completely inactive, enabling continued learning. This modification helps maintain stable, faster training and better gradient flow in deep networks. 

$$ LReLU(z) = max(0.01*z, z) $$

![](https://pbs.twimg.com/media/GFFUmhuaQAABXAc.jpg)

## **Parametric ReLU**
Parametric Rectified Linear Unit (PReLU) is a neural network activation function that improves upon ReLU by introducing a learnable parameter α for negative inputs. It adapts the negative slope during training, allowing the model to avoid "dead neurons" and optimize performance. 

$$ PReLU(z) = max(α*z, z) $$

![](https://media.geeksforgeeks.org/wp-content/uploads/20250129164914387404/parametric-relu.PNG)

## **ELU**
The Exponential Linear Unit (ELU) is a neural network activation function that improves upon ReLU by enabling faster learning and higher classification accuracy, particularly in deep networks. It solves the "dying ReLU" problem by allowing negative values with a smooth, exponential curve, bringing activations closer to zero-mean. 

$$
f(x) = 
\begin{cases} 
x, & x > 0 \\
\alpha(e^x - 1), & x \leq 0 
\end{cases}
$$

![](https://media.geeksforgeeks.org/wp-content/uploads/20250603154601294973/Elu_Activation_Function.png)

## **Softmax**
Softmax Activation Function converts a vector of raw neural network outputs (logits) into a probability distribution, where each value is between 0 and 1 and all values sum to 1. It is primarily used in the output layer for multi-class classification, representing the probability of each class.

$$ σ(z_{i}) = \frac{e^{z_{i}}}{\sum_{k=1}^{n} e^{z_{k}}} $$
