# **Loss functions**
A loss function in deep learning is a mathematical function that quantifies the difference between a model's predicted output and the actual target value (ground truth). It serves as a measure of the model's performance and, during training, acts as a guide for optimization algorithms (like gradient descent) to adjust the model's parameters (weights and biases) to minimize this error.

## Loss function vs cost function

| Feature | Loss Function | Cost Function |
| :--- | :--- | :--- |
| **Scope** | Refers to a **single** training example. | Refers to the **entire** training set (or a mini-batch). |
| **Purpose** | Measures the error between the predicted value and the actual label for one data point. | Measures the average error across all data points to evaluate the model's overall performance. |
| **Formula Context** | Calculated as $L(\hat{y}, y)$. | Calculated as the average of all losses, e.g., $J = \frac{1}{n} \sum L_i$. |
| **Optimization** | Provides the immediate "penalty" for a specific mistake. | The value that optimization algorithms (like Gradient Descent) seek to minimize. |
| **When to Use?** | Used during the forward pass to compute the error for each individual prediction. | Used during the backward pass (backpropagation) to update weights and biases based on the total error. |

## **Regression cost functions**

- ### **MSE (Mean squared error)**

Mean Squared Error (MSE) is a fundamental concept in statistics and machine learning playing a crucial role in assessing the accuracy of the predictive models. It is a parameter to calculate the accuracy of the model. It measures the average squared difference between predicted values and the actual values in the dataset.

**Advantage**
- It is always differentiable
- Has only one *Global* minima

**Disadvantage**
- Can be affected by outliers
- Not in same units (beacuse of squaring)

The cost function in MSE:
$$ C(θ₁, θ₂) = \sum^{n}_{i=1} \frac{(y-ŷ)²}{n} $$
$$ n = number\ of\ data\ points$$

- ### **MAE (Mean absolute error)**

Mean Absolute Error calculates the average difference between the calculated values and actual values. It is also known as scale-dependent accuracy as it calculates error in observations taken on the same scale used to predict the accuracy of the machine learning model.

**Advantage**
- It is less likely to be affected by outliers
- In same units

**Disadvantage**
- it is not differential at (0, 0)
- Time consuming

The cost function in MSE:

$$ C(θ₁, θ₂) = \sum^{n}_{i=1} \frac{|y-ŷ|}{n} $$
$$ n = number\ of\ data\ points$$

- ### **RMSE (Root mean squared error)**

Root mean square is defined as the quadratic mean or a subset of the generalized mean with an exponent of 2. To put it another way, the square root of the entire sum of squares of each data value in an observation is calculated using the root mean square formula. It can be interpreted as a changing function based on an integral of the squares of the values that are instantaneous in a cycle. It is used here to compute the square root of the arithmetic mean of the square of the function that describes the continuous waveform. It is abbreviated as RMS.

**Advantage**
- It is always differentiable
- In same units

**Disadvantage**
- Can be affected by outliers

$$ C(θ₁, θ₂) = \sqrt{MSE} $$
$$ OR $$
$$ C(θ₁, θ₂) = \sqrt{\sum^{n}_{i=1} \frac{(y-ŷ)²}{n}} $$

- ### **Huber Loss**

Huber Loss acts as a hybrid between Mean Squared Error (MSE) and Mean Absolute Error (MAE). It is designed to be robust to outliers while remaining differentiable. It behaves like MSE when the error is small (controlled by a threshold parameter $\delta$) and like MAE when the error is large, effectively getting the "best of both worlds."

**Advantage**
- Robust to outliers (like MAE)
- Differentiable at all points, including the origin (like MSE)
- Balanced approach to optimization

**Disadvantage**
- Introduces an additional hyperparameter ($\delta$) that needs tuning
- More computationally complex than MSE or MAE

$$
L_{δ}(y, ŷ) = 
\begin{cases} 
\frac{(y-ŷ)^{2}}{n}, & |y-ŷ| > δ \\
δ(\frac{|y-ŷ| - \frac{δ}{2}}{n}), & otherwise 
\end{cases}
$$

$$ C(θ₁, θ₂) = \frac{1}{n} (\sum^{n}_{i=1} L_{δ}(y, ŷ)) $$

## **Classification cost functions**

- ### **Binary Cross Entropy (BCE)**

Binary Cross Entropy is used for binary classification problems where the target variable has only two classes (0 and 1). It measures the difference between the actual class label and the predicted probability given by the model. It penalizes confident but wrong predictions heavily.

**Advantage**
- Works perfectly for binary classification
- Smooth and differentiable
- Strong penalty for wrong confident predictions

**Disadvantage**
- Sensitive to outliers or mislabeled data
- Requires predicted probabilities (not raw class labels)

The cost function in Binary Cross Entropy:

$$
C(θ) = -\frac{1}{n} \sum^{n}_{i=1} \left[ y_i \log(ŷ_i) + (1 - y_i)\log(1 - ŷ_i) \right]
$$

$$
n = number\ of\ data\ points
$$


- ### **Categorical Cross Entropy (CCE)**

Categorical Cross Entropy is used for multi-class classification problems where the target variable has more than two classes and is one-hot encoded. It compares the predicted probability distribution (usually from Softmax) with the actual distribution.

**Advantage**
- Ideal for multi-class problems
- Differentiable and works well with Softmax
- Strong theoretical foundation in information theory

**Disadvantage**
- Requires one-hot encoding of labels
- Computationally heavier for large number of classes

The cost function in Categorical Cross Entropy:

$$
C(θ) = -\frac{1}{n} \sum^{n}_{i=1} \sum^{k}_{j=1} y_{ij} \log(ŷ_{ij})
$$

$$
n = number\ of\ data\ points
$$
$$
k = number\ of\ classes
$$


- ### **Sparse Categorical Cross Entropy (SCCE)**

Sparse Categorical Cross Entropy is similar to Categorical Cross Entropy but is used when class labels are provided as integers instead of one-hot encoded vectors. It internally performs the same operation but is more memory efficient.

**Advantage**
- No need for one-hot encoding
- More memory efficient
- Suitable for large datasets

**Disadvantage**
- Slightly less intuitive than one-hot representation
- Still computationally heavy for very large class counts

The cost function in Sparse Categorical Cross Entropy:

$$
C(θ) = -\frac{1}{n} \sum^{n}_{i=1} \log(ŷ_{i, y_i})
$$

$$
n = number\ of\ data\ points
$$


- ### **Hinge Loss**

Hinge Loss is primarily used in Support Vector Machines (SVM). It is used for binary classification and focuses on maximizing the margin between classes. It penalizes predictions that are not only wrong but also those that are correct but not confident enough.

**Advantage**
- Encourages large margin separation
- Robust to small prediction errors
- Works well with SVM

**Disadvantage**
- Not probabilistic (does not output probabilities)
- Not differentiable at the hinge point

The cost function in Hinge Loss:

$$
C(θ) = \frac{1}{n} \sum^{n}_{i=1} \max(0, 1 - y_i f(x_i))
$$

$$
n = number\ of\ data\ points
$$
