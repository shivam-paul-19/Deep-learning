# **Perceptron** 
A Perceptron is the **simplest type of Artificial Neural Network**.

It is:
- A single neuron model
- Used for binary classification
- The foundation of modern neural networks

It was introduced by Frank Rosenblatt in 1958.

![](https://media.geeksforgeeks.org/wp-content/uploads/20251209120638608023/bhu.webp)

## 2. Structure of a Perceptron

A perceptron consists of:

- Inputs (x1, x2, x3, ..., xn)
- Weights (w1, w2, w3, ..., wn)
- Bias (b)
- Weighted sum
- Activation function
- Output (y)

### Mathematical Representation:

Weighted Sum:

$$ z = w1x1 + w2x2 + ... + wnxn + b $$
$$ \text{Output: } y = activation(z) $$

## 3. Activation Function in Perceptron

The classical perceptron uses the **Step Function**:

$$ \text{If } z >= 0 → Output = 1 $$
$$ \text{If } z < 0 → Output = 0 $$

This makes it suitable for **binary classification problems**.

## 4. How Perceptron Learns

The perceptron updates weights using a simple rule:

$$ w = w + λ × (y - ŷ) × input $$
$$ λ = \text{Learning rate} $$
$$ y = \text{actual value} $$
$$ ŷ = \text{predicted value} $$

This process is repeated until:
- The model correctly classifies all training data
OR
- It reaches maximum iterations

This is called the **Perceptron Learning Rule**.

## 5. Limitation of Perceptron

The perceptron can only solve:

- Linearly separable problems

It cannot solve:
- XOR problem
- Non-linear classification problems

This limitation led to:
- Multi-layer Perceptron (MLP)
- Deep Neural Networks

## 6. Why Perceptron is Important

Even though simple, it:

- Introduced the concept of weighted inputs
- Introduced learning via weight updates
- Became the foundation of modern neural networks

All deep learning models today are just advanced versions of stacked perceptrons.
