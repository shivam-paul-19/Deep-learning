# **Optimizers**
Optimizers in deep learning are algorithms that reduce training errors by updating neural network weights and biases based on loss function gradients. They navigate the optimization landscape to minimize loss, with key types including SGD (Stochastic Gradient Descent), Momentum, RMSprop, and Adam. Adam is frequently used for rapid, stable convergence.

## **Gradient descent**
Gradient Descent is the foundational optimization algorithm used to train machine learning and deep learning models. It functions by iteratively adjusting model parameters (weights and biases) to minimize a loss function, which measures the error between predicted and actual outcomes.

## **Stochastic Gradient Descent (SGD)**
In its "pure" form, SGD updates model parameters after seeing exactly one randomly selected training example. For every single data point, the model calculates the error, finds the gradient, and updates the weights immediately. It only needs to load one data point at a time, making it ideal for datasets too large for RAM. 

## **Mini batch SGD**
This is the industry standard for deep learning. It strikes a balance between pure SGD and full-batch gradient descent by updating weights based on a small subset (mini-batch) of data. You divide your dataset into small batches, calculate the average gradient for all points in that batch, and then perform a single weight update.

## **SGD with momentum**
While Mini-Batch SGD is efficient, it suffers from "oscillations"—the weights tend to zig-zag back and forth across a valley rather than racing down the center. SGD with Momentum solves this by adding a "physics" element to the optimization process.

Instead of updating the weights using just the current gradient, we maintain a velocity variable ($v$).

$$ \text{Smoothening the path between a1 and a2} $$
$$ v_{t-1} = a_{1} $$
$$ v_{t} = β * v_{t-1} + (1 - β) * a_{2} $$

$$ β = \text{A hyperparameter, rangin from 0 to 1, the more it's towards 1, the control is in a1. And vice versa} $$

## **Adagrad (Adaptive Gradient Algorithm)**
Adagrad (Adaptive Gradient Algorithm) is an optimization method that adjusts the learning rate for each parameter during training. Unlike standard gradient descent with a fixed rate, Adagrad uses past gradients to scale updates making it effective for sparse data and varying feature magnitudes.

$$ ɳ' = \frac{ɳ}{\sqrt(α_{t} + ε)} $$
$$ ɳ' = \text{New learning rate} $$
$$ ε = \text{A small value that prevent denominator to become 0} $$

$$ α_{t} = \sum^{t}_{i=1} (\frac{∂ʎ}{∂w_{t}}) $$

## **Adadelta**
Adadelta is an advanced, adaptive gradient descent optimization algorithm that extends AdaGrad to enhance training speed and robustness. It addresses the issue of rapidly diminishing learning rates by using a moving average of squared gradients, eliminating the need for manual learning rate tuning.

$$ ɳ' = \frac{ɳ}{\sqrt(sd_{w_{t}} + ε)} $$
$$ ɳ' = \text{New learning rate} $$
$$ ε = \text{A small value that prevent denominator to become 0} $$

$$ sd_{w_{t}} = β * sd_{w_{t-1}} + (1 - β) * (\frac{∂ʎ}{∂w_{t-1}})^2 $$

## **Adam**
Adam (Adaptive Moment Estimation) is a widely used, computationally efficient optimization algorithm in deep learning that updates network weights iteratively based on training data. By combining the advantages of AdaGrad (handling sparse gradients) and RMSProp (handling non-stationary objectives), Adam computes individual adaptive learning rates for different parameters.

First moment (mean of gradients):

$$ m_t = β_1 m_{t-1} + (1 - β_1) \frac{∂ʎ}{∂w_t} $$

Second moment (uncentered variance):

$$ v_t = β_2 v_{t-1} + (1 - β_2) \left( \frac{∂ʎ}{∂w_t} \right)^2 $$

Bias correction:

$$ \hat{m_t} = \frac{m_t}{1 - β_1^t} $$
$$ \hat{v_t} = \frac{v_t}{1 - β_2^t} $$

Weight update rule:

$$ w_{t+1} = w_t - ɳ \frac{\hat{m_t}}{\sqrt{\hat{v_t}} + ε} $$

Where:

$$ β_1 = \text{Exponential decay rate for first moment (commonly 0.9)} $$
$$ β_2 = \text{Exponential decay rate for second moment (commonly 0.999)} $$
$$ ε = \text{A small value that prevent denominator to become 0} $$

## Vanishing Gradient problem
The vanishing gradient problem occurs in deep neural networks when gradients used for weight updates become extremely small during backpropagation. As gradients are multiplied across many layers, they shrink exponentially, causing early layers to learn slowly or stop learning altogether. This often results in poor, slow convergence and is primarily caused by activation functions like $sigmoid$ or $tanh$

## Exploding Gradient problem
The exploding gradient problem occurs in deep neural networks when, during backpropagation, gradients accumulate and grow exponentially, becoming excessively large. This leads to unstable training, huge weight updates, and loss divergence (often resulting in NaN values), commonly appearing in RNNs or deep, poorly initialized networks.
