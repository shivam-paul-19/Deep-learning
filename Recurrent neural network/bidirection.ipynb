# **Bidirectional Recurrent Neural Networks (BiRNN)**

A **Bidirectional Recurrent Neural Network (BiRNN)** is an extension of a standard RNN that processes sequence data **in both directions**:

- **Forward pass:** left → right
- **Backward pass:** right → left

Instead of learning context only from past tokens, a BiRNN learns from **both past and future context simultaneously**.

![](https://media.geeksforgeeks.org/wp-content/uploads/20230302163012/Bidirectional-Recurrent-Neural-Network-2.png)

### Standard RNN

```
x1 → x2 → x3 → x4
↓    ↓    ↓    ↓
h1   h2   h3   h4
```

Each hidden state only depends on **previous inputs**.

### Bidirectional RNN

```
Forward direction:
x1 → x2 → x3 → x4
↓    ↓    ↓    ↓
h1f  h2f  h3f  h4f

Backward direction:
x1 ← x2 ← x3 ← x4
↓    ↓    ↓    ↓
h1b  h2b  h3b  h4b
```

Final representation at timestep `t`:


$ h_{t} = [h_{t_{{forward}}}, h_{t_{backward}}] $


(`,` denotes **concatenation**)

So the model learns information from:

- **past context**
- **future context**

## Why is Bidirectional RNN Needed?

Standard RNNs only consider **previous information**.

However, in many tasks the **future words also give meaning**.

### Example

Sentence:

```
He went to the bank to withdraw money.
```

The word **"bank"** could mean:

- river bank
- financial bank

The word **"withdraw"** appears later and helps clarify the meaning.

A normal RNN sees:

```
He → went → to → the → bank
```

But a **BiRNN also sees future context**:

```
bank ← withdraw ← money
```

Thus the representation becomes richer.

### Typical Applications

Bidirectional RNNs are commonly used in:

- Named Entity Recognition (NER)
- Speech recognition
- Machine translation
- Sentiment analysis
- Part-of-speech tagging

## Mathematical Representation

### Forward RNN

x_t = input at time t
h_t = hidden state
W = weight matrices
b = bias
σ = activation function

The forward hidden state is computed as:

$$h_{t}^{f} = σ(W_{xh}^{f} * x_{t} + W_{hh}^{f} * h_(t-1)^{f} + b_{f})$$

Where:

$$
h_{t}^{f} → \text{forward hidden state}
W_{xh}^{f} → \text{input-to-hidden weights}
W_{hh}^{f} → \text{hidden-to-hidden weights}
$$

This processes the sequence **from start to end**.

### Backward RNN

The backward hidden state is computed as:

$$h_{t}^{b} = σ(W_{xh}^{b} * x_{t} + W_{hh}^{b} * h_(t+1)^{b} + b_{b})$$

Where:
$$
h_{t}^{b} → \text{backward hidden state}
$$

This processes the sequence **from end to start**.

### Combined Hidden State

The final hidden representation at timestep `t` is:

$$h_{t} = [h_{t}^{f} ; h_{t}^{b}]$$

or sometimes:
$$h_{t} = h_{t}^{f} + h_{t}^{b}$$


depending on implementation.

## Output Computation

If a prediction is made at every timestep:

$$ y_{t} = softmax(W_{yh} * h_{t} + b_{y}) $$

Since `h_t` contains information from both directions, the prediction uses **complete contextual information**.

## Backpropagation in Bidirectional RNN

Training uses **Backpropagation Through Time (BPTT)** in both directions.

Two independent gradient flows occur:

### Forward Network Gradient

∂L/∂W_f = Σ_t (∂L/∂h_t^f × ∂h_t^f/∂W_f)

Gradients propagate:

```
t → t-1 → t-2 → ...
```

### Backward Network Gradient

∂L/∂W_b = Σ_t (∂L/∂h_t^b × ∂h_t^b/∂W_b)

Gradients propagate:

```
t → t+1 → t+2 → ...
```

Thus two gradient chains are computed independently.

## Key Insight

A Bidirectional RNN effectively maintains **two separate RNNs**:

- Forward RNN
- Backward RNN

Both learn different contextual representations, which are then **combined to improve prediction quality**.
