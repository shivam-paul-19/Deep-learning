# **Transformers**
Transformer is a neural network architecture used for performing machine learning tasks particularly in natural language processing (NLP) and computer vision. In 2017 Vaswani et al. published a paper " Attention is All You Need" in which the transformers architecture was introduced. The article explores the architecture, workings and applications of transformers.

![](https://media.geeksforgeeks.org/wp-content/uploads/20251210153206327851/transformers.webp)

## Working

The **Transformers** use a series of Encoders and Decoders to perform NLP tasks, better than classic *Encoder-decoder*. Unlike them, it doesn't take the input one by one, instead, all the words are sent together.

- #### **Self attention**

    Self Attention mechanism is used to find the *Contexual vectors* from the embedded vectors of each word.

    Embedded vector (`[x, y, z]`) *-- Self attention -->* Contexual vector (`[x', y', z']`)\
    Where, `x'` means the word's relevance with the first word, `y'` shows how much context it takes from second word

    It does with help of the help of three vectors:
    
    - ***Query (Q)***: What it is looking for?
    - ***Key (K)***: What it has to offer?
    - ***Value (V)***: What value/information it holds?

- #### **Multi head attention**

    Instead of one attention mechanism, transformers use multiple attention heads running in parallel. Each head captures different relationships or patterns in the data, enriching the model’s understanding.

- #### **Positional Encoding**

    Unlike RNNs, transformers lack an inherent understanding of word order since they process data in parallel. To solve this problem Positional Encodings are added to token embeddings providing information about the position of each token within a sequence.

    It can be problematic as order matters in most phrases.\
    If a model treats **Tiger ate Deer** and **Deer ate Tiger** as equal then it would be a blunder. So the model needs to know that *Tiger* came first in the sentence. 
    For this we make *Positional vectors* and add them in the embeddings.

- #### **Masked multi head attention**

    During padding, which is done to make the length of vectors equal, sometimes **0s** affect the calcualtions a lot. In small amount, they are harmless but suppose, the largest sentence has 20 words in it and one sentence has 3 words, the 17 **0s** will definitely affect the values and calculation.\
    To deal with this, we need *Masked multi head attention*.

All this components work together to form a transformer model.

```[]
# Encoder
Encoder: 
  Input → Embeddings → Positional Encoding → Multi-Head Self-Attention → FFNN
  (Repeat N times)

# Decoder
Decoder:
  Output → Embeddings → Positional Encoding → Masked Multi-Head Self-Attention 
         → FFNN
  (Repeat N times)
```
