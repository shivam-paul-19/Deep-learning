# **Self Attention**
Self-attention is a mechanism in Transformer models that allows each token in a sequence to dynamically weigh the importance of all other tokens, creating context-aware representations. By calculating scaled dot-product attention between Queries (Q), Keys (K), and Values (V) derived from the same input, it captures long-range dependencies efficiently in parallel.

![](https://media.geeksforgeeks.org/wp-content/uploads/20250506114140815865/frame_3053.webp)

## Working

The Self attention mechanism comrpises of three vectors:
- **Query (Q)**: the question a token is asking. "What context do I need to understand myself?" It represents the current token's information need.
- **Key (K)**: the label/index that each token exposes to others. "Here's what I'm about." It's used to match against incoming queries.
- **Value (V)**: the actual content a token contributes. Once a query decides a key is relevant, the corresponding value is what actually gets passed along.

### **Initialising Q, K, V**
Firstly, the vectors are initialised for each vectors with the help of $W$ vectors, they contain weight which are learned during training. They are multiplied with the embedded vector ($E$) of each word.

$$ Q_{word_{1}} = W_{Q} ⋅ (E_{word_{1}})^{T} $$
$$ K_{word_{1}} = W_{K} ⋅ (E_{word_{1}})^{T} $$
$$ V_{word_{1}} = W_{V} ⋅ (E_{word_{1}})^{T} $$

![](https://jalammar.github.io/images/t/self-attention-matrix-calculation.png)

### **Calculating Attention score**
After getting the $Q$ and $K$ vector, and attention socre is calculated for every pair of token using $Q$ and $K$ vectors of them.

$$ AS(Q_{word_{1}}, K_{word_{2}}) = Q_{word_{1}} ⋅ (K_{word_{2}})^{T} $$
$$ AS(Q_{word_{1}}, K_{word_{3}}) = Q_{word_{1}} ⋅ (K_{word_{3}})^{T} $$
$$ AS(Q_{word_{1}}, K_{word_{4}}) = Q_{word_{1}} ⋅ (K_{word_{4}})^{T} $$
$$ AS = \text{Attention score} $$

### **Scaling**
It is th process of scalling down the attention score by dividing them with $\sqrt{D_{k}}$, where $D_{x} = \text{Dimension of x vector}$.
#### Why it is done?
To avoid the *Exploding gradient problem* and *Softmax Saturation*.

![](https://jalammar.github.io/images/t/self-attention-matrix-calculation-2.png)

### **Get Attention weights**
After calculating the *Attention scores* and scaling them, we calculate attention weights by applying **softmax**.

$$ AT_{word_{1}} = Softmax(\text{All scaled ASs of word 1}) $$
$$ AT = \text{Attention weights} $$

### **Get the contextual vector**
After getting the attention weights, we calculate the *Sum of weight multiplied with V vector*.

$$ Output(word_{1}) = \sum^{n}_{i=1} AT_{word_{1}}[i] * V_{word_{i}} $$

The $Output(word_{1})$ is the contextual vector of $word_{1}$ after passing through self attention mechanism.

# **Multi head attention**
In multi headed attention mechanism, multiple attention heads are used in parallel which allows the model to focus on different parts of the input sequence simultaneously. This approach increases model's ability to capture various relationships between words in the sequence.

![](https://jalammar.github.io/images/t/transformer_attention_heads_qkv.png)

Then all the matrices of multiple attention values are concatenated and multipled with a $W_{0}$ matrix to make it one matrix. As the feed forward neural network is expecting just one vector/matrix, not multiple.

![](https://jalammar.github.io/images/t/transformer_attention_heads_weight_matrix_o.png)
