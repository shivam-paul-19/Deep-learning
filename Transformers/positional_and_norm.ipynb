# **Positional encoding**
Positional encoding injects word order information into transformer models, which process tokens in parallel rather than sequentially. It adds a unique vector to each input embedding to represent its position, enabling the model to distinguish between words based on their location, such as in "Tiger eats deer" vs. "Deer eats tiger".

# Sinusoidal Positional Encoding
Sinusoidal Positional Encoding is a technique used in Transformer models to inject information about the order of tokens (words, characters, etc.) in a sequence. Since Transformers process all tokens in a sequence simultaneously rather than sequentially (like Recurrent Neural Networks), they lack inherent knowledge of token order. This method, introduced in the 2017 paper "Attention Is All You Need," uses sine and cosine functions of varying frequencies to generate a unique, fixed vector for each position, which is then added directly to the input token embeddings.

Formula:

$$ PE_{pos, 2i} = sin(\frac{pos}{10000^{\frac{2i}{d_{model}}}}) $$
$$ PE_{pos, 2i+1} = cos(\frac{pos}{10000^{\frac{2i}{d_{model}}}}) $$

$$ where, $$
$$ p = \text{Position} $$
$$ i = \text{dimension} $$
$$ d_{model} = \text{dimensionality of model} $$

### *Example*
> Example sentence: "The cat sat"

Suppose the embeddings are:

- **The** -> `[0.1, 0.2, 0.3, 0.4]`
- **cat** -> `[0.5, 0.6, 0.7, 0.8]`
- **sat** -> `[0.9, 1.0, 1.1, 1.2]`

Here,
$ d_{model} = 4 $

$ P(0, 0) = sin(\frac{0}{10000^{\frac{0}{4}}}) = 0 $\
$ P(0, 1) = cos(\frac{0}{10000^{\frac{1}{4}}}) = 1 $\
$ P(0, 2) = sin(\frac{0}{10000^{\frac{2}{4}}}) = 0 $\
$ P(0, 3) = cos(\frac{0}{10000^{\frac{3}{4}}}) = 1 $

...

$ P(1, 0) = sin(\frac{1}{10000^{\frac{0}{4}}}) = 0.8415 $\
$ P(1, 1) = cos(\frac{1}{10000^{\frac{1}{4}}}) = 0.5403 $\
$ P(1, 2) = sin(\frac{1}{10000^{\frac{2}{4}}}) = 0.01 $\
$ P(1, 3) = cos(\frac{1}{10000^{\frac{3}{4}}}) = 0.99995 $

...

$ P(2, 0) = sin(\frac{2}{10000^{\frac{0}{4}}}) = 0.9093 $\
$ P(2, 1) = cos(\frac{2}{10000^{\frac{1}{4}}}) = 0.9801 $\
$ P(2, 2) = sin(\frac{2}{10000^{\frac{2}{4}}}) = 0.02 $\
$ P(2, 3) = cos(\frac{2}{10000^{\frac{3}{4}}}) = 0.999998 $

...

The positional encoding will be:

- **The** -> `[0.0, 1.0, 0.0, 1.0]`
- **cat** -> `[0.8415, 0.5403, 0.01, 0.99995]`
- **sat** -> `[0.9093, 0.9801, 0.02, 0.999998]`

These positional encodings are added with the embeddings before giving to feed forwad neural network

# **Normalisation**
Normalization in deep learning is a data preparation and architectural technique that scales input features and hidden layer activations to a common range (typically zero mean and unit variance). It stabilizes training, accelerates convergence, and prevents features with larger scales from dominating the model. Common techniques include **Batch Normalization (BN)**, **Layer Normalization (LN)**

## **Batch Normalisation**
Batch Normalization is used to reduce the problem of internal covariate shift in neural networks. It works by normalizing the data within each mini-batch. This means it calculates the mean and variance of data in a batch and then adjusts the values so that they have similar range. After that it scales and shifts the values so that model learn effectively.

Need for it:
- Solves the problem of internal covariate shift.
- Makes training faster and more stable.
- Allows use of higher learning rates.
- Helps avoid vanishing or exploding gradients.
- Can act like a regularizer sometimes reduce the need for dropout.

Formula:

$$ x_{i}' = \frac{x_{i} - μ}{\sqrt{σ^{2} + ε}}  $$
$$ Where,  $$
$$ x_{i}' = \text{new x point}  $$
$$ x_{i} = \text{original x point}  $$
$$ σ² = \text{Variance}  $$
$$ ε = \text{A hyperparameter to ensure the denominator is always non-zero} $$

## **Layer Normalisation**
Layer Normalization stabilizes and accelerates the training process in deep learning. In typical neural networks, activations of each layer can vary drastically which leads to issues like exploding or vanishing gradients which slow down training. Layer Normalization addresses this by normalizing the output of each layer which helps in ensuring that the activations stay within a stable range.

It works by normalizing the input to each neuron such that the mean activation becomes 0 and the variance becomes 1. Unlike Batch Normalization which normalizes over the batch i.e across all samples in the batch, it normalizes over the features for each individual data point.

![](https://cdn.sanity.io/images/vr8gru94/production/409b7645d3bdc19d267f6a6bea3bbf75f70636f7-800x535.png)

In this nomralisation, we also do **Scaling** and **Shifting**, Scaling (γ) and shifting (β) in Layer Normalization allow the model to learn the optimal distribution for activations, ensuring that normalization doesn't restrict the network's representation power. While normalization fixes the mean to 0 and variance to 1, learnable parameters restore necessary flexibility, stability, and representational capacity to the network.

$$ y_{i} = x_{i}'*γ_{i} + β_{i} $$
$$ x_{i}' = \text{new x point after normalising} $$
