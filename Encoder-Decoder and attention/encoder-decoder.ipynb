# **Encoder Decoder**
Encoder-decoder models in deep learning are neural network architectures designed to process input sequences and generate output sequences, often of different lengths. The encoder compresses input data (text, image, audio) into a compact, fixed-dimensional "context vector" (latent space), while the decoder takes this representation to generate the target output, commonly used in machine translation, text summarization, and image captioning.

## Working

### 1. **Word embeddings**
The text is first embedded into vectors and two vectors are added, `<SOS>` and `<EOS>` which marks **Start of sentence** and **End of sentence**.

### 2. **Encoding mechanism**
The words are sent one by one in each time frame, for eg:

- t1 -> 'once'
- t2 -> 'upon'
- t3 -> 'a'
- t4 -> 'time'

The last layer sends a `Context vector` to the decoder for the sequence prediction. 

### 3. **Decoder mechanism**
The decoder gets the `context vector` and an input as `<SOS>` and applies a `Softmax` activation to find the predicted embeddings for the word.

- $t_{n}$ (`<SOS>` goes as input here) -> first word embedding -> comes as 'il'
- $t_{n + 1}$ (il goes as input here) -> second word embedding -> comes as 'était'

and this goes till `<EOS>`

## Limitations
Encoder-decoder models, particularly in neural machine translation and sequence modeling, suffer from limitations including information bottlenecks with long sequences, high training complexity, and potential information loss when compressing input into a single fixed-size context vector. They often struggle with long-term dependencies, lack explicit alignment, and may produce "structural stereotypes" in image tasks.
To reduce this, we need **Attention mechanism**.
