# **Attention Mechanism**
The Attention Mechanism is a technique that allows models to focus on the most important parts of input data when making predictions. It assigns different weights to different elements hence helping the model prioritize relevant information instead of treating all inputs equally. It forms the foundation of advanced models like Transformers and BERT and is widely used in Natural Language Processing (NLP).

![](https://media.geeksforgeeks.org/wp-content/uploads/20251029093447681289/string_constant_pool_5.webp)

## Why it is needed?

In a standard encoder-decoder setup, the encoder processes words one by one. Each step updates a "hidden state" that is supposed to carry all the information from previous steps. 

- **The Chain of Memory**: For the word at position 10 to be remembered, its information must be successfully passed through the hidden states of positions 11, 12, 13, and so on, until the final state (the context vector).
- **Vanishing Gradients**: As the sequence gets longer, the mathematical signal from the earliest words becomes increasingly faint (the vanishing gradient problem).
- **The Result**: By the time the encoder reaches the end of a long sentence, the final context vector is heavily skewed toward the last few words it "saw." The decoder then starts its work with a summary that is mostly about the end of the input, leading to errors at the beginning of the output. 

## Working

- ### **Encoder**
    - It uses a series of bidirectional RNN/LSTM.
    - Here each LSTM network also produces an output $h_{n}$.
    - The multiple $h$ outputs are feeded in an ANN to calculate the *Context vector*. Here, unlike simple encoder, the context vector has the information of not only the current word but also the previous words that came.

- ### **Decoder**
    - As the Decoder precicts each output, it know sees two things.
    - It sees the word which was predicted before it, and also the context that came from the encoder, which is needed to know the importance of words that previously came.

**Example:**
Suppose an English to french translation is going on,

English sentence: "I ate a pizza but it wasn't good."

Now, the words are constantly predicted by the decoder like:

- *J'ai mangé une pizza mais cela n'était pas bonne*.
- When the decoder was translating "good", it could be "bon" or "bonne" (based on gender), for this it had to look for the context.
- The attention mechanism will put light on the word "pizza" which is feminine in french so the decoder can get a context that "good" is referred to what?
