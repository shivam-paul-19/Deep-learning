# **Deep Learning Notes**

## 1. What is Deep Learning?

Deep Learning (DL) is a subset of Machine Learning that uses **Artificial Neural Networks (ANNs)** with multiple layers to learn patterns from data.

In simple terms:
- ML = Teaching a machine with rules + features
- DL = Machine learns the features by itself

![](https://media.geeksforgeeks.org/wp-content/uploads/20250702125756538577/_What-is-Deep-Learning_1.webp)

### Core Idea:
Deep Learning tries to mimic how the human brain works using:
- Neurons
- Layers (Input → Hidden → Output)
- Weights and biases
- Activation functions

The word **"Deep"** comes from having **many hidden layers** in a neural network.

## 2. Why Deep Learning Exists When Machine Learning Already Exists?

Good question. ML was powerful... but it had limitations.

### Problems with Traditional Machine Learning:

#### 1. Manual Feature Engineering
- In ML, humans must manually decide which features are important.
- Example: For image recognition, you must extract edges, shapes, textures manually.
- This is time-consuming and requires domain expertise.

#### 2. Poor Performance on Unstructured Data
ML struggles with:
- Images
- Audio
- Natural language
- Videos

Because these data types are:
- High dimensional
- Complex
- Non-linear

#### 3. Scalability Issues
Traditional ML models do not scale well with:
- Massive datasets
- Extremely complex patterns

### What Deep Learning Solves:

- Automatically learns features from raw data
- Performs better with large datasets
- Handles unstructured data efficiently
- Captures highly complex, non-linear relationships

In short:
> ML needs guidance.  
> DL learns representations by itself.
